# Description
- count: The total number of non-null (or non-NA) observations in each column.
- mean: The average value of all non-null observations in each column. 
- std: This stands for standard deviation, which measures the dispersion of a dataset relative to its mean. 
- min: The smallest value in each column.
- 25%: Also known as the first quartile, the value below which 25% of the observations fall. For column 1, 25% of all values are below 6 million, and for column 2, 25% of all values are below 7 million.
- 50%: Also known as the median or the second quartile, this is the middle value in the dataset.
- 75%: Also known as the third quartile, the value below which 75% of the observations fall.
- max: The largest value in each column.
- These values are represented in scientific notation.
